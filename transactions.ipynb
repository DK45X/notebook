```python
import pandas as pd
import random
from faker import Faker
import numpy as np

fake = Faker()
Faker.seed(42)
random.seed(42)

def introducir_errores(data):
    for i in range(len(data)):
        if random.random() < 0.1:  # 10% de errores
            data[i] = None
        elif random.random() < 0.1:
            data[i] = fake.word()
    return data

data = {
    'ID': [fake.uuid4() for _ in range(5000)],
    'Nombre': introducir_errores([fake.name() for _ in range(5000)]),
    'Teléfono': introducir_errores([fake.phone_number() for _ in range(5000)]),
    'Correo': introducir_errores([fake.email() for _ in range(5000)]),
    'Dirección': introducir_errores([fake.address() for _ in range(5000)]),
    'Último_Contacto': introducir_errores([fake.date_this_year() for _ in range(5000)]),
    'Edad': introducir_errores([random.choice([None, fake.random_int(min=18, max=90)]) for _ in range(5000)]),
    'Ingreso': introducir_errores([random.choice([None, fake.random_int(min=10000, max=1000000)]) for _ in range(5000)])
}

df = pd.DataFrame(data)
print(df.head())

# Guardar el DataFrame en un archivo CSV
df.to_csv('contactabilidad_datos_sucios.csv', index=False)
print("Conjunto de datos creado y guardado en 'contactabilidad_datos_sucios.csv'")
```
